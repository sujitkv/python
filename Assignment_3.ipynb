Q.1 Why are functions advantageous to have in your programs?

Ans. Functions in Python enhance code organization, readability, reusability, and maintenance, allowing modular design and efficient collaboration in software development.

Q.2 When does the code in a function run: when it's specified or when it's called?

Ans. The function execution occurs when it is called/invoke either by name or by providing arguments, not when specified/defined.

Q.3 What statement creates a function?

Ans. In Python, the def statement is used to create a function. after the "def" keyword function name come on place, a pair of parentheses containing optional parameters, and a colon. The indented block of code that follows defines the function's behavior. The return statement, if used, specifies the value that the function will return when it's called.

def function_name(parameters):
    # Function code or body
    # ...
    return result
    
Q.4 What is the difference between a function and a function call?

Ans. A function is a code block with a purpose, defined using def. A function call is the act of executing a function with specific arguments, making it run.

Q5. How many global scopes are there in a Python program? How many local scopes?

Ans. Python has one global scope per program, but local scopes are created within functions, potentially many.

Q6. What happens to variables in a local scope when the function call returns?

Ans. Local scope variables cease to exist (get destroyed) after the function call returns.

Q7. What is the concept of a return value? Is it possible to have a return value in an expression?

Ans. A return value is the result a function sends back. It can be used in expressions, e.g., result = function()

Q8. If a function does not have a return statement, what is the return value of a call to that function?

Ans. If no return statement, Python returns None by default for that function call.

Q9. How do you make a function variable refer to the global variable?

Ans. To make a function use a global variable, declare it as global within the function, like global_var = global_value.

Q10. What is the data type of None?

Ans. The data type of None is NoneType.

Q11. What does the sentence import areallyourpetsnamederic do?

Ans. It imports a module named "areallyourpetsnamederic," allowing access to its functions and variables.

Q12. If you had a bacon() feature in a spam module, what would you call it after importing spam?

Ans. After importing spam, you would call bacon() as spam.bacon().

Q13. What can you do to save a program from crashing if it encounters an error?

Ans. Use error handling with try-except blocks to gracefully handle errors and prevent crashes.

Q14. What is the purpose of the try clause? What is the purpose of the except clause?

Ans. The try clause attempts risky code, while the except clause handles exceptions, allowing controlled error management.
